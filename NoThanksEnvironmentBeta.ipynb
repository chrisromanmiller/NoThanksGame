```python
import numpy as np

class Env():
    def __init__(self):
        self.done = False
        
        self.unrevealed_cards = [1 for i in range(36)] #unplayed_cards[i] = 1 if i has not been played. 0-2 is garbage
        self.unrevealed_cards[0], self.unrevealed_cards[1], self.unrevealed_cards[2] = 0,0,0   
        self.current_card = None
        self.reveal_card()
        
        self.num_tokens = 1 


        
    def reveal_card(self):
        card = np.random.choice([i for i in range(3,36) if self.unrevealed_cards[i] ==1])
        self.unrevealed_cards[card] = 0
        self.current_card = card
        
        
    def _take(self):
        if sum(self.unrevealed_cards) == 9:
            self.done = True
        self.num_tokens = 0
        self.reveal_card()

        
        
    def _pass(self):
        self.num_tokens +=1
        
    def check_state(self,verbose= False):
        if verbose:
            print("There are {0} tokens and the curent card is {1}. {2} cards remain unrevealed".format(self.num_tokens, self.current_card, sum(self.unrevealed_cards)-9 ))
            if self.done:
                print("The game is over")


        return self.num_tokens, self.current_card
            
class Player():
    def __init__(self, starting_tokens = 5, player_type = "Human"):
        self.num_tokens = starting_tokens
        self.cards_posessed = [0 for i in range(36)]
        self.player_type = player_type
        

        
    def get_action(self, env):
        if self.num_tokens == 0:
            return "take"
        if self.player_type == "Random":
            return np.random.choice(["take", "pass"], p= [.2,.8])
        
        if self.player_type == "Human":
            return "take"

    def _pass(self):
        self.num_tokens -=1
        
    def _take(self, card, num_tokens):
        self.num_tokens += num_tokens
        self.cards_posessed[card] = 1    
        
    def status_report(self):
        print("{0} tokens and cards {1}".format(self.num_tokens, [i for i in range(36) if self.cards_posessed[i] ==1]))
        
class Game():
    def __init__(self, number_players):
        self.number_players = number_players
        self.player_list = [Player(starting_tokens= 10, player_type="Random") for i in range(number_players)]
        self.env = Env()
        self.turn_number = 0
        
    def advance(self):
        player = self.player_list[self.turn_number % self.number_players]
        action = player.get_action(self.env)
        
        tokens, card = self.env.check_state()
        if action == "take":
            player._take(card,tokens)
            self.env._take()
            
        if action == "pass":
            player._pass()
            self.env._pass()
            
            
        self.turn_number +=1
        
    def status_report(self):
        self.env.check_state(verbose = True)
        for player in self.player_list:
            player.status_report()
        
        
    def play_game(self):
        while self.env.done == False:
            self.advance()
            self.status_report()
            print()
        
```

```python
game = Game(5)
game.play_game()
```
